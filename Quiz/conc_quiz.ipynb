## 문제 2

### 토지 파일에서 단어를 입력받아 입력받은 단어가 포함된 문장을 출력해주는 함수를 작성하시오
### 그 후 출력된 결과값을 conc_res라는 변수 명에 저장하시오
    (입력 받을 단어 예시 : 강포수)

```python
def conc(word):
    
    # 토지2 파일 경로 저장
    txt = "E:\Programming\python\창회선배스터디\창회선배 자료\텍스트자료\토지2.txt"

    # 파일 읽기
    with open(txt, "r", encoding='UTF-8') as filetoread:
        fileread = filetoread.read()
        # 온점 기준으로 나누기
        orig_txt = fileread.split('.')

    # 원하는 단어의 문장 찾기
    res = []
    for x in orig_txt:
        if word in x:
            res.append(x)

    return(res)

conc_res = conc(input('단어'))

print(conc_res)
```

```python
import os

# 간단 문장 용례 추출기 전문

def simple_conc(example_file_path, search_word, result_file_path, result_file_name):
    txt = "E:\Programming\python\창회선배스터디\창회선배 자료\텍스트자료\토지2.txt"

    text = []
    with open(txt, "r", encoding='UTF-8') as filetoread:
        fileread = filetoread.readlines()
        for txt_by_line in fileread:
            text.append(txt_by_line)

    divided_txt = []
    for x in text:
        k = x.split('.')
        divided_txt.append(k)
    divided_txt = sum(divided_txt, [])

    res = []
    for x in divided_txt:
        if search_word in x:
            res.append(x)

    res_path = result_file_path + result_file_name + ".txt"
    result = open(res_path, 'w', encoding='cp949')
    k = 0
    for i in res:
        result.write(str(k+1) + ' => ' + str(i).strip() + '\n')
        k += 1
    result.close()
    os.startfile(res_path)

    return(res)
```
